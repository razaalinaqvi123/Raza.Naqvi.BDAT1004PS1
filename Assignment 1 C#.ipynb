```

// Implement function triangleArea(a,b,c) that takes as input the lengths of the 3
// sides of a triangle and returns the area of the triangle. By Heron's formula, the area
// of a triangle with side lengths a, b, and c is s(s − a)(s −b)(s −c), where
// s = (a + b + c)/2.
//  triangleArea(2,2,2)
// 1.7320508075688772

static double triangleArea(int a, int b, int c) {
    Console.WriteLine("a: "+a);
    Console.WriteLine("b: "+b);
    Console.WriteLine("c: "+c);
    
    double s = (a + b + c) / 2.0d; 
    double area = Math.Sqrt(s*(s-a) * (s - b) * (s - c)); 
    
    return area;
}

Console.WriteLine("Please enter the value of a b and c for getting the area of the triangle");

a= Console.ReadLine();
b= Console.ReadLine();
c= Console.ReadLine();

double x = triangleArea(a,b,c);
// double x = triangleArea(5,4,3);
Console.WriteLine("The area of the triangle is: "+x);
```

```
// Question 5 C#
// a. Write a function inside(x,y,x1,y1,x2,y2) that returns True or False
// dependingonwhetherthepoint(x,y)liesintherectanglewithlowerleft
// corner (x1,y1) and upper right corner (x2,y2).
// >>> inside(1,1,0,0,2,3)
// True
// >>> inside(-1,-1,0,0,2,3)
// False
// b. Use function inside()from part a. to write an expression that tests whether
// the point (1,1) lies in both of the following rectangles: one with lower left
// corner (0.3, 0.5) and upper right corner (1.1, 0.7) and the other with lower
// left corner (0.5, 0.2) and upper right corner (1.1, 2)
    
double inside(double x, double y, double x1, double y1, double x2, double y2)
{
    if (x > x1 && x < x2 && y > y1 && y < y2)
    {
        Console.WriteLine("True");
        return 1;
    }
    else
    {
        Console.WriteLine("False");
        return 0;
    }
}

inside(1, 1, 0, 0, 2, 3);
inside(-1, -1, 0, 0, 2, 3);

// PART B
if (inside(1, 1, 0.3, 0.5, 1.1, 0.7) == 1 && inside(1, 1, 0.5, 0.2, 1.1, 2) == 1)
{
    Console.WriteLine("The point is inside both rectangles");
} else
{
    Console.WriteLine("The point isn't inside both rectangles");
}    
```
